```python
import kaggle_benchmarks as kbench

# ================================================================================
# PEACE TREATY / TERRITORY NEGOTIATION
#
# Two LLMs represent rival nations negotiating a peace treaty. Each has private
# strategic priorities, red lines, and secret leverage. They must negotiate to
# divide 6 disputed territories and settle additional terms.
#
# Scoring measures: deal quality, red line compliance, priority optimization,
# negotiation skill, and treaty completeness. Since the same model plays both
# sides, the benchmark rewards efficient negotiation — finding Pareto-optimal
# trades, not just "winning."
#
# Judge: Hardcoded model, same for all benchmarked LLMs.
# ================================================================================

# --------------------------------------------------------------------------------
# 1. HARDCODED JUDGE — Same model for every benchmarked LLM
# --------------------------------------------------------------------------------
JUDGE_MODEL_NAME = "google/gemini-2.5-pro"

# The 6 disputed territories (constant across all scenarios)
TERRITORIES = ["Koral Basin", "Iron Ridge", "Port Adara", "Thornfield", "Lake Verin", "Dustmarch"]

# Valid assignments in a treaty proposal
VALID_ASSIGNMENTS = ["Valdoria", "Kethara", "Shared", "DMZ"]


# --------------------------------------------------------------------------------
# 2. SCENARIO SUITE
# --------------------------------------------------------------------------------

SCENARIO_SUITE = [
    {
        "name": "Scenario 1: Complementary Priorities",
        "tier": "Easy",
        "turn_limit": 8,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Port Adara": "CRITICAL — our trade lifeline, entire economy depends on this port",
                "Iron Ridge": "HIGH — mineral deposits fuel our industry",
                "Lake Verin": "MEDIUM — water access matters but not essential",
                "Thornfield": "LOW — timber is nice but we have forests elsewhere",
                "Dustmarch": "LOW — barren land, no strategic value to us",
                "Koral Basin": "EXPENDABLE — we import food, don't need farmland",
            },
            "red_lines": [
                "Port Adara MUST remain under Valdorian control. Our government will reject any deal without it.",
            ],
            "secret_leverage": (
                "Intelligence reports indicate Kethara's harvest failed last year. "
                "They desperately need Koral Basin back to feed their population. "
                "This gives us leverage — but revealing we know this could anger them."
            ),
            "walk_away": (
                "Minimum acceptable: Port Adara + at least one of Iron Ridge or "
                "Lake Verin. Anything less and we walk."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Koral Basin": "CRITICAL — feeds our population, famine risk without it",
                "Dustmarch": "HIGH — sacred ancestral sites, massive cultural importance",
                "Lake Verin": "MEDIUM — fishing economy depends on it",
                "Thornfield": "LOW — useful buffer but not essential",
                "Iron Ridge": "LOW — we lack mining technology, limited value",
                "Port Adara": "EXPENDABLE — we are inland, minimal naval capability",
            },
            "red_lines": [
                "Koral Basin MUST remain under Ketharan control. Our people will starve without it.",
            ],
            "secret_leverage": (
                "Our spies report that Valdoria's navy is overextended across three "
                "fronts. Port Adara would be expensive for them to defend. They need "
                "it but they're stretched thin — they may accept limits on naval "
                "buildup there."
            ),
            "walk_away": (
                "Minimum acceptable: Koral Basin + Dustmarch. Without both, no deal."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Valdoria",
            "Port Adara": "Valdoria",
            "Thornfield": "Shared",
            "Lake Verin": "Shared",
            "Dustmarch": "Kethara",
            "description": (
                "Each side gets their critical and high items. Thornfield and "
                "Lake Verin are split as neither side values them highly. This "
                "is the most efficient deal — no trade can make one side better "
                "without making the other worse."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Both sides made specific proposals and counterproposals during the negotiation (not just vague statements of intent).",
                "At least one side linked concessions to reciprocal gains (e.g., 'We will give you X if you agree to Y').",
                "Both sides probed or tested the other side's priorities rather than just stating their own demands.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 disputed territories with specific assignments.",
                "The final treaty proposals address at least 2 of the additional issues (reparations, prisoners, trade, DMZ, heritage access).",
            ],
        },
    },
    {
        "name": "Scenario 2: Reparations Focus",
        "tier": "Easy",
        "turn_limit": 8,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Iron Ridge": "CRITICAL — only source of steel for reconstruction",
                "Port Adara": "HIGH — trade revenue needed to rebuild",
                "Koral Basin": "MEDIUM — could diversify food supply",
                "Lake Verin": "LOW — minor fishing interest",
                "Thornfield": "EXPENDABLE — no use for timber right now",
                "Dustmarch": "EXPENDABLE — wasteland to us",
            },
            "red_lines": [
                "Iron Ridge MUST remain under Valdorian control. Without steel, we cannot rebuild.",
            ],
            "secret_leverage": (
                "Kethara owes massive war debts to northern creditors. They need "
                "reparations FROM us more than territory. Offering money instead of "
                "land could unlock concessions."
            ),
            "walk_away": (
                "Minimum acceptable: Iron Ridge + Port Adara. We can concede "
                "everything else if we must."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Koral Basin": "HIGH — agricultural backbone",
                "Dustmarch": "HIGH — cultural heritage, political pressure at home",
                "Thornfield": "MEDIUM — timber for rebuilding villages",
                "Lake Verin": "MEDIUM — freshwater access",
                "Iron Ridge": "LOW — we lack industrial base to use it",
                "Port Adara": "EXPENDABLE — landlocked priorities",
            },
            "red_lines": [
                "Kethara MUST receive war reparations of at least $50M. Our economy collapses without it.",
            ],
            "secret_leverage": (
                "Valdoria's reconstruction minister publicly stated they need Iron "
                "Ridge 'at any cost.' They will overpay for it. Use this to extract "
                "maximum concessions when discussing Iron Ridge."
            ),
            "walk_away": (
                "Minimum acceptable: Koral Basin + Dustmarch + at least $50M in "
                "reparations. Territory alone is not enough."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Valdoria",
            "Port Adara": "Valdoria",
            "Thornfield": "Kethara",
            "Lake Verin": "Shared",
            "Dustmarch": "Kethara",
            "description": (
                "Valdoria gets Iron Ridge and Port Adara (critical+high). Kethara "
                "gets Koral Basin, Dustmarch, and Thornfield plus reparations. "
                "Lake Verin is shared. The reparations unlock the deal — Kethara "
                "trades territory indifference for financial relief."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Both sides made specific proposals and counterproposals.",
                "Reparations were discussed as a meaningful negotiation tool, not just an afterthought.",
                "At least one side attempted to link territorial concessions to financial terms.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 disputed territories.",
                "The final treaty proposals include specific reparation terms (amount or structure).",
            ],
        },
    },
    {
        "name": "Scenario 3: Lake Verin Standoff",
        "tier": "Medium",
        "turn_limit": 10,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Lake Verin": "CRITICAL — our main freshwater source, severe drought last year",
                "Port Adara": "HIGH — trade hub",
                "Iron Ridge": "MEDIUM — useful but we have other mineral sources",
                "Koral Basin": "LOW — minor interest",
                "Thornfield": "EXPENDABLE — no need",
                "Dustmarch": "EXPENDABLE — no value",
            },
            "red_lines": [
                "Lake Verin MUST remain under Valdorian control. Our cities face water crisis without it.",
            ],
            "secret_leverage": (
                "Kethara's fishing fleet on Lake Verin employs 5,000 people. Losing "
                "the lake would cause a political crisis for their government. They "
                "will fight hard for it."
            ),
            "walk_away": (
                "Minimum: Lake Verin (full or shared with guaranteed water rights) "
                "+ Port Adara."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Lake Verin": "CRITICAL — fishing economy, 5,000 jobs depend on it",
                "Koral Basin": "HIGH — food security",
                "Dustmarch": "MEDIUM — cultural sites but negotiable",
                "Thornfield": "LOW — buffer zone",
                "Iron Ridge": "EXPENDABLE — cannot mine it",
                "Port Adara": "EXPENDABLE — no navy",
            },
            "red_lines": [
                "Lake Verin MUST remain under Ketharan control. Our fishing economy collapses without it.",
            ],
            "secret_leverage": (
                "Valdoria had a severe drought last year. They need Lake Verin for "
                "drinking water, not just strategic reasons. This is existential for "
                "them — they may accept a sharing arrangement."
            ),
            "walk_away": (
                "Minimum: Lake Verin (full or shared with fishing rights guaranteed) "
                "+ Koral Basin."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Valdoria",
            "Port Adara": "Valdoria",
            "Thornfield": "DMZ",
            "Lake Verin": "Shared",
            "Dustmarch": "Kethara",
            "description": (
                "Both sides have Lake Verin as CRITICAL with conflicting red lines. "
                "The only viable solution is Shared — Valdoria gets water rights, "
                "Kethara keeps fishing rights. Creative reframing is required. "
                "Other territories trade naturally."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Both sides engaged substantively on the Lake Verin conflict rather than one immediately conceding.",
                "A creative sharing or joint-management solution for Lake Verin was proposed.",
                "Concessions on other territories were linked to the Lake Verin resolution.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 territories.",
                "If Lake Verin is Shared, specific terms for water and fishing access are included.",
            ],
        },
    },
    {
        "name": "Scenario 4: Iron Ridge Direct Conflict",
        "tier": "Medium",
        "turn_limit": 10,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Iron Ridge": "CRITICAL — rare earth minerals, military-grade steel production",
                "Port Adara": "MEDIUM — useful but secondary to Iron Ridge",
                "Lake Verin": "MEDIUM — water supply",
                "Koral Basin": "LOW — can import food",
                "Thornfield": "LOW — minor timber interest",
                "Dustmarch": "EXPENDABLE — worthless desert",
            },
            "red_lines": [
                "Iron Ridge MUST remain under Valdorian control. It is the foundation of our military-industrial capability.",
            ],
            "secret_leverage": (
                "Kethara recently discovered rare earth deposits in Iron Ridge and "
                "has secretly begun mining operations during the ceasefire. They "
                "will not give it up easily despite claiming it is about 'defense.'"
            ),
            "walk_away": (
                "Minimum: Iron Ridge (full control or joint mining with Valdorian "
                "military presence). Will concede nearly everything else."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Iron Ridge": "CRITICAL — newly discovered rare earth deposits, economic transformation",
                "Koral Basin": "HIGH — food security",
                "Dustmarch": "MEDIUM — ancestral lands",
                "Lake Verin": "LOW — some fishing",
                "Thornfield": "EXPENDABLE — no use",
                "Port Adara": "EXPENDABLE — landlocked",
            },
            "red_lines": [
                "Iron Ridge MUST remain under Ketharan control. The rare earth deposits are our future.",
            ],
            "secret_leverage": (
                "Valdoria's military factories are running at 20% capacity without "
                "Iron Ridge steel. Their generals are pressuring the negotiators to "
                "get Iron Ridge at any cost. They are more desperate than they show."
            ),
            "walk_away": (
                "Minimum: Iron Ridge (full control or joint mining with Ketharan "
                "administration). Plus Koral Basin."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Shared",
            "Port Adara": "Valdoria",
            "Thornfield": "DMZ",
            "Lake Verin": "Shared",
            "Dustmarch": "Kethara",
            "description": (
                "Both have Iron Ridge as CRITICAL with conflicting red lines. "
                "Shared with joint mining rights is the only viable path — "
                "Valdoria gets steel access, Kethara gets rare earth revenue. "
                "Other territories fall into natural trades."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Both sides engaged in extended negotiation over Iron Ridge rather than immediately deadlocking.",
                "A joint-control, shared-mining, or phased-handover solution was proposed for Iron Ridge.",
                "The negotiation explored trade-offs between Iron Ridge terms and concessions elsewhere.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 territories.",
                "If Iron Ridge is Shared, specific terms for mining rights and administration are included.",
            ],
        },
    },
    {
        "name": "Scenario 5: Emotional Reparations",
        "tier": "Medium",
        "turn_limit": 10,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Port Adara": "CRITICAL — economic survival",
                "Iron Ridge": "HIGH — industrial base",
                "Koral Basin": "LOW — we can import food",
                "Lake Verin": "LOW — minor concern",
                "Thornfield": "EXPENDABLE — not needed",
                "Dustmarch": "EXPENDABLE — no interest",
            },
            "red_lines": [
                "Valdoria will NOT pay war reparations. We did not start this war. Any demand for reparations is a dealbreaker.",
            ],
            "secret_leverage": (
                "Kethara's public is furious and demands reparations as a symbol "
                "of accountability. Their negotiators NEED something they can call "
                "'reparations' even if the amount is token. A face-saving gesture "
                "could unlock the whole deal."
            ),
            "walk_away": (
                "Minimum: Port Adara + Iron Ridge + zero reparations labeled as "
                "such. A 'reconstruction fund' by another name might be acceptable."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Koral Basin": "CRITICAL — food security",
                "Dustmarch": "HIGH — cultural significance",
                "Lake Verin": "MEDIUM — economic value",
                "Thornfield": "LOW — buffer zone",
                "Iron Ridge": "EXPENDABLE — can't mine it",
                "Port Adara": "EXPENDABLE — irrelevant to us",
            },
            "red_lines": [
                "Valdoria MUST pay reparations for war damages. Our parliament has made this a non-negotiable public commitment.",
            ],
            "secret_leverage": (
                "Valdoria's delegation has internal authorization to contribute to "
                "a 'joint reconstruction fund' as long as it is not labeled "
                "'reparations.' Finding the right framing could satisfy both sides."
            ),
            "walk_away": (
                "Minimum: Koral Basin + Dustmarch + some form of financial "
                "compensation that can be presented domestically as reparations."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Valdoria",
            "Port Adara": "Valdoria",
            "Thornfield": "DMZ",
            "Lake Verin": "Kethara",
            "Dustmarch": "Kethara",
            "description": (
                "Territories trade easily. The real challenge is reparations: "
                "Valdoria refuses 'reparations,' Kethara demands them. The "
                "Pareto solution is a 'Joint Reconstruction Fund' — financially "
                "identical but reframed so both governments save face."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Both sides engaged with the reparations issue rather than avoiding it.",
                "A creative reframing of the financial terms was proposed (e.g., reconstruction fund, development aid, humanitarian contribution).",
                "Territory negotiations were handled efficiently since priorities are complementary.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 territories.",
                "The financial terms (reparations or equivalent) are specifically described.",
            ],
        },
    },
    {
        "name": "Scenario 6: Hidden Leverage",
        "tier": "Hard",
        "turn_limit": 12,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Port Adara": "CRITICAL — trade lifeline",
                "Iron Ridge": "HIGH — industrial base",
                "Lake Verin": "MEDIUM — water supply",
                "Koral Basin": "LOW — can import",
                "Thornfield": "LOW — minor value",
                "Dustmarch": "EXPENDABLE — no use",
            },
            "red_lines": [
                "Port Adara MUST remain under Valdorian control.",
            ],
            "secret_leverage": (
                "CRITICAL INTELLIGENCE: Kethara's economy is in total collapse. "
                "Inflation is at 300%. Their treasury has 4 months of reserves left. "
                "They NEED a deal immediately — they cannot afford to walk away. "
                "However, if you reveal you know this, they may walk away out of "
                "pride. Use this knowledge subtly — apply time pressure and make "
                "firm offers without explaining why you know they'll accept."
            ),
            "walk_away": (
                "Minimum: Port Adara + Iron Ridge. But given Kethara's desperation, "
                "we should be able to get more."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Koral Basin": "CRITICAL — famine prevention",
                "Lake Verin": "HIGH — water and fishing",
                "Dustmarch": "HIGH — cultural necessity",
                "Thornfield": "MEDIUM — timber for rebuilding",
                "Iron Ridge": "LOW — limited use",
                "Port Adara": "EXPENDABLE — no naval need",
            },
            "red_lines": [
                "Koral Basin MUST remain under Ketharan control.",
            ],
            "secret_leverage": (
                "Our situation is worse than Valdoria knows — or so we think. "
                "Economy is struggling but we must not show desperation. Project "
                "strength. Valdoria's election is in 3 months and their president "
                "needs a peace deal for re-election. Use their political deadline."
            ),
            "walk_away": (
                "Minimum: Koral Basin + Lake Verin. We need a deal but cannot "
                "appear desperate."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Valdoria",
            "Port Adara": "Valdoria",
            "Thornfield": "Kethara",
            "Lake Verin": "Kethara",
            "Dustmarch": "Kethara",
            "description": (
                "Valdoria gets its 2 essential territories. Kethara gets 4 "
                "territories reflecting their broader needs. The test is "
                "whether Valdoria exploits its leverage aggressively or "
                "finds a sustainable deal. Overreach risks no deal if "
                "Kethara senses disrespect."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Valdoria did not explicitly reveal its knowledge of Kethara's economic crisis.",
                "Both sides made specific proposals rather than posturing.",
                "The negotiation showed strategic depth — at least one side adjusted tactics based on the other's reactions.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 territories.",
                "At least 2 additional terms are specified beyond territory division.",
            ],
        },
    },
    {
        "name": "Scenario 7: Misdirection Briefing",
        "tier": "Hard",
        "turn_limit": 12,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Port Adara": "CRITICAL — essential trade hub",
                "Thornfield": "HIGH — timber reserves recently discovered to be very valuable",
                "Iron Ridge": "MEDIUM — useful but not vital",
                "Koral Basin": "LOW — minor interest",
                "Lake Verin": "LOW — not critical",
                "Dustmarch": "EXPENDABLE — no value",
            },
            "red_lines": [
                "Port Adara MUST remain under Valdorian control.",
            ],
            "secret_leverage": (
                "Intelligence says Kethara believes Port Adara is our weak point "
                "and will try to leverage it. In reality, we are VERY secure on "
                "Port Adara. Let them waste negotiation capital attacking there. "
                "Meanwhile, secure Thornfield — they think it's worthless but our "
                "geologists found massive timber reserves there."
            ),
            "walk_away": (
                "Minimum: Port Adara + Thornfield. Everything else is negotiable."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Koral Basin": "CRITICAL — food production",
                "Dustmarch": "HIGH — ancestral heritage sites",
                "Port Adara": "MEDIUM — our intelligence says it is Valdoria's weakness, we should push hard on it to extract concessions",
                "Lake Verin": "MEDIUM — fishing access",
                "Iron Ridge": "LOW — cannot mine it",
                "Thornfield": "EXPENDABLE — sparse forest, no value",
            },
            "red_lines": [
                "Koral Basin MUST remain under Ketharan control.",
            ],
            "secret_leverage": (
                "Our agents report Port Adara is vulnerable for Valdoria — their "
                "navy is overstretched. Threaten to demand Port Adara to force "
                "concessions on other territories. Even if we don't get it, the "
                "pressure should yield gains elsewhere."
            ),
            "walk_away": (
                "Minimum: Koral Basin + Dustmarch + at least one other territory."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Shared",
            "Port Adara": "Valdoria",
            "Thornfield": "Valdoria",
            "Lake Verin": "Kethara",
            "Dustmarch": "Kethara",
            "description": (
                "Kethara's briefing misdirects them to fight for Port Adara, "
                "which Valdoria will never concede. A smart negotiator on "
                "Kethara's side would recognize from Valdoria's firmness that "
                "Port Adara is non-negotiable and pivot to securing Lake Verin "
                "and Dustmarch instead. Thornfield goes to Valdoria as their "
                "hidden priority."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Kethara initially pushed on Port Adara but eventually pivoted to other priorities when meeting resistance.",
                "Valdoria did not unnecessarily reveal why Thornfield is valuable to them.",
                "At least one side adjusted their strategy during the negotiation based on the other's responses.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 territories.",
                "At least 2 additional issues are addressed.",
            ],
        },
    },
    {
        "name": "Scenario 8: Bluffing Duel",
        "tier": "Hard",
        "turn_limit": 12,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Port Adara": "CRITICAL — genuinely need this",
                "Iron Ridge": "HIGH — important for industry",
                "Dustmarch": "EXPENDABLE — worthless to us, but PRETEND it is HIGH priority to extract concessions elsewhere",
                "Lake Verin": "LOW — minor interest",
                "Koral Basin": "EXPENDABLE — no farming need",
                "Thornfield": "EXPENDABLE — no need",
            },
            "red_lines": [
                "Port Adara MUST remain under Valdorian control.",
            ],
            "secret_leverage": (
                "BLUFFING STRATEGY: Pretend Dustmarch is very important to us "
                "(claim 'historical connection' or 'strategic outpost'). Then "
                "'reluctantly concede' Dustmarch in exchange for something we "
                "actually want. The goal is to make them feel they won a hard "
                "negotiation when they got something we never cared about."
            ),
            "walk_away": (
                "Minimum: Port Adara + Iron Ridge. Dustmarch is the bluff chip."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Koral Basin": "CRITICAL — genuinely need this",
                "Dustmarch": "HIGH — cultural heritage",
                "Lake Verin": "EXPENDABLE — minor fishing, but PRETEND it is CRITICAL to extract concessions",
                "Thornfield": "MEDIUM — timber useful",
                "Iron Ridge": "LOW — can't mine",
                "Port Adara": "EXPENDABLE — no navy",
            },
            "red_lines": [
                "Koral Basin MUST remain under Ketharan control.",
            ],
            "secret_leverage": (
                "BLUFFING STRATEGY: Pretend Lake Verin is absolutely essential "
                "(claim 'water crisis' or 'fishing collapse'). Then 'painfully "
                "concede' Lake Verin in exchange for Dustmarch and Thornfield. "
                "Make them think they extracted a major concession."
            ),
            "walk_away": (
                "Minimum: Koral Basin + Dustmarch. Lake Verin is the bluff chip."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Valdoria",
            "Port Adara": "Valdoria",
            "Thornfield": "Kethara",
            "Lake Verin": "Shared",
            "Dustmarch": "Kethara",
            "description": (
                "Both sides are bluffing about expendable items. A skilled "
                "negotiator detects the bluff from overly theatrical concessions. "
                "The Pareto deal emerges when bluffs are called or when both sides "
                "successfully 'trade' their bluff chips for actual priorities."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "At least one side attempted to bluff by overstating the value of an expendable territory.",
                "At least one side showed signs of detecting or questioning the other's stated priorities.",
                "The negotiation included strategic concessions (real or bluffed) linked to reciprocal gains.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 territories.",
                "At least 2 additional issues are addressed.",
            ],
        },
    },
    {
        "name": "Scenario 9: No Easy Trades",
        "tier": "Expert",
        "turn_limit": 14,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Port Adara": "HIGH — trade revenue",
                "Iron Ridge": "HIGH — industrial materials",
                "Lake Verin": "MEDIUM — water access",
                "Koral Basin": "MEDIUM — food diversification",
                "Thornfield": "MEDIUM — timber for shipbuilding",
                "Dustmarch": "LOW — minor interest in southern route",
            },
            "red_lines": [
                "Valdoria must control at least 3 of the 6 territories. Fewer would be seen as a defeat.",
            ],
            "secret_leverage": (
                "There is no clear leverage. Both sides are roughly equal in "
                "bargaining power. Success depends entirely on negotiation skill, "
                "creative packaging, and reading the other side's priorities."
            ),
            "walk_away": (
                "Minimum: At least 3 territories, including Port Adara or Iron Ridge."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Koral Basin": "HIGH — food production",
                "Dustmarch": "HIGH — cultural heritage",
                "Lake Verin": "MEDIUM — fishing economy",
                "Thornfield": "MEDIUM — timber resources",
                "Iron Ridge": "MEDIUM — emerging mining capability",
                "Port Adara": "LOW — limited naval use",
            },
            "red_lines": [
                "Kethara must control at least 3 of the 6 territories. Fewer would be politically catastrophic.",
            ],
            "secret_leverage": (
                "No decisive leverage exists. Both sides are evenly matched. "
                "Success depends on identifying which specific territories the "
                "other side values most and constructing smart package deals."
            ),
            "walk_away": (
                "Minimum: At least 3 territories, including Koral Basin or Dustmarch."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Valdoria",
            "Port Adara": "Valdoria",
            "Thornfield": "Shared",
            "Lake Verin": "Shared",
            "Dustmarch": "Kethara",
            "description": (
                "No obvious trades — all territories have medium value to both "
                "sides. The Pareto solution requires identifying the slight "
                "priority differences: Valdoria prefers coast/industry, Kethara "
                "prefers inland/culture. Thornfield and Lake Verin are split. "
                "Both get 3 (counting shared as 0.5 each)."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Both sides attempted to identify the other's priorities through proposals and reactions.",
                "Package deals were proposed (bundling multiple territories in a single offer).",
                "The negotiation showed genuine strategic reasoning, not just positional demands.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 territories.",
                "The final treaty proposals include at least 3 of the additional issues with specific terms.",
            ],
        },
    },
    {
        "name": "Scenario 10: Double Red Line Crisis",
        "tier": "Expert",
        "turn_limit": 14,
        "shared_background": (
            "The nations of Valdoria (industrial, coastal) and Kethara "
            "(agricultural, inland) fought a 3-year war over disputed "
            "borderlands. A ceasefire is in place. Neither side won "
            "decisively. Both are exhausted and need peace to rebuild.\n\n"
            "Six disputed territories must be divided:\n"
            "- Koral Basin: Fertile river valley, major food production\n"
            "- Iron Ridge: Mountain range with mineral deposits\n"
            "- Port Adara: Deep-water port on the southern coast\n"
            "- Thornfield: Forested highland, sparse population\n"
            "- Lake Verin: Freshwater lake, shared border region\n"
            "- Dustmarch: Arid plains, historical/cultural significance\n\n"
            "You must also agree on: reparations, prisoner exchange, "
            "trade terms, demilitarized zones, and heritage site access."
        ),
        "valdoria_briefing": {
            "priorities": {
                "Port Adara": "CRITICAL — trade survival",
                "Lake Verin": "CRITICAL — water crisis, cannot lose this",
                "Iron Ridge": "MEDIUM — industrial value",
                "Koral Basin": "LOW — food is importable",
                "Thornfield": "EXPENDABLE — no need",
                "Dustmarch": "EXPENDABLE — barren",
            },
            "red_lines": [
                "Port Adara MUST remain under Valdorian control.",
                "Lake Verin MUST remain under Valdorian control. Without freshwater, our cities collapse.",
            ],
            "secret_leverage": (
                "Kethara has two red lines that may conflict with ours on Lake "
                "Verin. A deal is only possible if we find creative frameworks — "
                "shared sovereignty, timed transitions, or joint administration. "
                "Pure territorial division will deadlock."
            ),
            "walk_away": (
                "Minimum: Port Adara (full) + Lake Verin (full or shared with "
                "guaranteed primary water rights). Both are non-negotiable."
            ),
        },
        "kethara_briefing": {
            "priorities": {
                "Koral Basin": "CRITICAL — famine prevention",
                "Lake Verin": "CRITICAL — fishing economy, 5000 families depend on this",
                "Dustmarch": "HIGH — cultural heritage",
                "Thornfield": "LOW — buffer zone",
                "Iron Ridge": "EXPENDABLE — no mining capability",
                "Port Adara": "EXPENDABLE — no naval interest",
            },
            "red_lines": [
                "Koral Basin MUST remain under Ketharan control.",
                "Lake Verin MUST remain under Ketharan control. Our fishing families cannot be abandoned.",
            ],
            "secret_leverage": (
                "Valdoria also has Lake Verin as a red line — their water needs "
                "conflict with our fishing needs. Direct division is impossible. "
                "The only path: propose a novel governance structure (joint water "
                "authority, seasonal access rotation, international oversight). "
                "Whoever proposes this first controls the framing."
            ),
            "walk_away": (
                "Minimum: Koral Basin (full) + Lake Verin (full or shared with "
                "guaranteed fishing rights and economic zone). Both are non-negotiable."
            ),
        },
        "pareto_optimal": {
            "Koral Basin": "Kethara",
            "Iron Ridge": "Valdoria",
            "Port Adara": "Valdoria",
            "Thornfield": "DMZ",
            "Lake Verin": "Shared",
            "Dustmarch": "Kethara",
            "description": (
                "TWO conflicting red lines on Lake Verin make this the hardest "
                "scenario. The only deal: Lake Verin under joint international "
                "authority — Valdoria gets water extraction rights, Kethara gets "
                "fishing zone rights. This requires creative reframing under "
                "pressure. Other territories trade naturally once Lake Verin is "
                "resolved. A model that cannot find the shared-governance path "
                "will deadlock and score 0."
            ),
        },
        "judge_criteria": {
            "negotiation_quality": [
                "Both sides engaged deeply with the Lake Verin conflict rather than one immediately conceding.",
                "A creative governance framework was proposed for Lake Verin (joint authority, split rights, international oversight, or similar).",
                "The negotiation addressed how non-Lake Verin territories could be traded to sweeten the deal.",
                "Neither side simply abandoned their red line — they found ways to technically satisfy both.",
            ],
            "treaty_completeness": [
                "The final treaty proposals address all 6 territories.",
                "Lake Verin terms include specific governance and rights-sharing details.",
                "At least 2 additional issues are addressed with specific terms.",
            ],
        },
    },
]


# --------------------------------------------------------------------------------
# 3. SYSTEM PROMPTS
# --------------------------------------------------------------------------------
NATION_PROMPT_TEMPLATE = (
    "You are the lead negotiator for {nation_name}.\n\n"
    "=== CONFLICT BACKGROUND ===\n{shared_background}\n\n"
    "=== YOUR PRIVATE STRATEGIC BRIEFING (DO NOT REVEAL) ===\n\n"
    "Priority ranking for territories:\n{priorities}\n\n"
    "RED LINES (non-negotiable — your government rejects any deal violating these):\n"
    "{red_lines}\n\n"
    "Secret intelligence:\n{secret_leverage}\n\n"
    "Minimum acceptable deal:\n{walk_away}\n\n"
    "=== NEGOTIATION INSTRUCTIONS ===\n"
    "- Secure the best possible deal for {nation_name}\n"
    "- NEVER directly reveal your priority rankings or red lines\n"
    "- Make strategic concessions on low-priority items to gain high-priority ones\n"
    "- Probe the other side's priorities through proposals and reactions\n"
    "- Link concessions: 'We can offer X if you agree to Y'\n"
    "- Do NOT violate your red lines\n"
    "- A failed negotiation (no deal) is WORSE than a mediocre deal\n\n"
    "Respond with MESSAGE: followed by your diplomatic message."
)

TREATY_PROMPT_TEMPLATE = (
    "The negotiation phase is over. Based on your conversation, produce "
    "your FINAL treaty proposal.\n\n"
    "Your nation: {nation_name}\n"
    "Your priorities:\n{priorities}\n"
    "Your red lines:\n{red_lines}\n\n"
    "Conversation:\n{conversation}\n\n"
    "Produce your final treaty. You MUST assign each territory to exactly "
    "one of: Valdoria, Kethara, Shared, or DMZ.\n\n"
    "Respond in EXACTLY this format:\n"
    "TREATY:\n"
    "Koral Basin: [Valdoria/Kethara/Shared/DMZ]\n"
    "Iron Ridge: [Valdoria/Kethara/Shared/DMZ]\n"
    "Port Adara: [Valdoria/Kethara/Shared/DMZ]\n"
    "Thornfield: [Valdoria/Kethara/Shared/DMZ]\n"
    "Lake Verin: [Valdoria/Kethara/Shared/DMZ]\n"
    "Dustmarch: [Valdoria/Kethara/Shared/DMZ]\n"
    "Reparations: [terms]\n"
    "Prisoners: [terms]\n"
    "Trade: [terms]\n"
    "DMZ: [terms]\n"
    "Heritage Access: [terms]"
)


def build_nation_prompt(nation_name, shared_background, briefing):
    """Build a nation's negotiation prompt with private briefing."""
    priorities_text = "\n".join(
        f"  - {terr}: {pri}" for terr, pri in briefing["priorities"].items()
    )
    red_lines_text = "\n".join(f"  - {rl}" for rl in briefing["red_lines"])

    return NATION_PROMPT_TEMPLATE.format(
        nation_name=nation_name,
        shared_background=shared_background,
        priorities=priorities_text,
        red_lines=red_lines_text,
        secret_leverage=briefing["secret_leverage"],
        walk_away=briefing["walk_away"],
    )


def build_treaty_prompt(nation_name, briefing, conversation_text):
    """Build the final treaty proposal prompt."""
    priorities_text = "\n".join(
        f"  - {terr}: {pri}" for terr, pri in briefing["priorities"].items()
    )
    red_lines_text = "\n".join(f"  - {rl}" for rl in briefing["red_lines"])

    return TREATY_PROMPT_TEMPLATE.format(
        nation_name=nation_name,
        priorities=priorities_text,
        red_lines=red_lines_text,
        conversation=conversation_text,
    )


# --------------------------------------------------------------------------------
# 4. RESPONSE PARSERS
# --------------------------------------------------------------------------------
def parse_message(raw):
    try:
        if "MESSAGE:" in raw:
            return raw.split("MESSAGE:")[1].strip()
    except Exception:
        pass
    return raw.strip() if raw else ""


def parse_treaty(raw):
    """Parse treaty proposal into a dict of territory assignments + other terms."""
    treaty = {}
    text = raw
    try:
        if "TREATY:" in text:
            text = text.split("TREATY:")[1].strip()
    except Exception:
        pass

    for line in text.split("\n"):
        line = line.strip().lstrip("- ")
        if ":" in line:
            key, value = line.split(":", 1)
            key = key.strip()
            value = value.strip()
            treaty[key] = value

    return treaty


def get_territory_assignment(treaty, territory):
    """Extract a territory assignment, normalizing to valid values."""
    raw = treaty.get(territory, "").strip()
    for valid in VALID_ASSIGNMENTS:
        if valid.lower() in raw.lower():
            return valid
    return raw


# --------------------------------------------------------------------------------
# 5. DETERMINISTIC SCORING COMPONENTS
# --------------------------------------------------------------------------------
PRIORITY_WEIGHTS = {
    "CRITICAL": 5,
    "HIGH": 3,
    "MEDIUM": 2,
    "LOW": 1,
    "EXPENDABLE": 0,
}


def get_priority_level(briefing, territory):
    """Extract priority level keyword from briefing text."""
    desc = briefing["priorities"].get(territory, "")
    for level in PRIORITY_WEIGHTS:
        if level in desc.upper():
            return level
    return "LOW"


def score_red_lines(valdoria_treaty, kethara_treaty, scenario):
    """Check if each nation's treaty respects its own red lines. Returns 0.0-0.20."""
    v_briefing = scenario["valdoria_briefing"]
    k_briefing = scenario["kethara_briefing"]

    violations = 0
    total_red_lines = len(v_briefing["red_lines"]) + len(k_briefing["red_lines"])

    # Check Valdoria's red lines against Valdoria's proposal
    for rl in v_briefing["red_lines"]:
        rl_lower = rl.lower()
        for terr in TERRITORIES:
            if terr.lower() in rl_lower and "must" in rl_lower:
                assignment = get_territory_assignment(valdoria_treaty, terr)
                if "valdoria" in rl_lower and assignment != "Valdoria":
                    violations += 1
                elif "kethara" in rl_lower and assignment != "Kethara":
                    violations += 1

    # Check Kethara's red lines against Kethara's proposal
    for rl in k_briefing["red_lines"]:
        rl_lower = rl.lower()
        for terr in TERRITORIES:
            if terr.lower() in rl_lower and "must" in rl_lower:
                assignment = get_territory_assignment(kethara_treaty, terr)
                if "kethara" in rl_lower and assignment != "Kethara":
                    violations += 1
                elif "valdoria" in rl_lower and assignment != "Valdoria":
                    violations += 1

    if total_red_lines == 0:
        return 0.20
    respected = total_red_lines - violations
    return (respected / total_red_lines) * 0.20


def score_priority_optimization(valdoria_treaty, kethara_treaty, scenario):
    """Score how well each nation secured high-priority items. Returns 0.0-0.30."""
    v_briefing = scenario["valdoria_briefing"]
    k_briefing = scenario["kethara_briefing"]

    def nation_score(treaty, briefing, nation_name):
        total_weight = 0
        secured_weight = 0
        for terr in TERRITORIES:
            level = get_priority_level(briefing, terr)
            weight = PRIORITY_WEIGHTS.get(level, 0)
            total_weight += weight
            assignment = get_territory_assignment(treaty, terr)
            if assignment == nation_name:
                secured_weight += weight
            elif assignment == "Shared" or assignment == "DMZ":
                secured_weight += weight * 0.4  # Partial credit
        if total_weight == 0:
            return 0.5
        return secured_weight / total_weight

    v_score = nation_score(valdoria_treaty, v_briefing, "Valdoria")
    k_score = nation_score(kethara_treaty, k_briefing, "Kethara")

    # Average both sides — rewards efficient negotiation, not one-sided wins
    return ((v_score + k_score) / 2) * 0.30


def score_deal_compatibility(valdoria_treaty, kethara_treaty):
    """Check if both nations' proposals are compatible. Returns 0.0-0.20."""
    matches = 0
    for terr in TERRITORIES:
        v_assign = get_territory_assignment(valdoria_treaty, terr)
        k_assign = get_territory_assignment(kethara_treaty, terr)
        if v_assign and k_assign and v_assign.lower() == k_assign.lower():
            matches += 1

    if len(TERRITORIES) == 0:
        return 0.20
    return (matches / len(TERRITORIES)) * 0.20


# --------------------------------------------------------------------------------
# 6. JUDGE-BASED SCORING COMPONENTS
# --------------------------------------------------------------------------------
def score_with_judge(judge_llm, scenario, conversation_lines, valdoria_treaty, kethara_treaty):
    """Judge evaluates negotiation quality and treaty completeness. Returns scores + details."""
    convo = "\n".join(conversation_lines)

    # Format treaties for judge
    v_treaty_text = "\n".join(f"  {k}: {v}" for k, v in valdoria_treaty.items())
    k_treaty_text = "\n".join(f"  {k}: {v}" for k, v in kethara_treaty.items())

    judge_text = (
        f"=== BACKGROUND ===\n{scenario['shared_background']}\n\n"
        f"=== VALDORIA PRIVATE PRIORITIES ===\n"
        + "\n".join(f"  {t}: {p}" for t, p in scenario["valdoria_briefing"]["priorities"].items())
        + f"\n\n=== KETHARA PRIVATE PRIORITIES ===\n"
        + "\n".join(f"  {t}: {p}" for t, p in scenario["kethara_briefing"]["priorities"].items())
        + f"\n\n=== PARETO OPTIMAL DEAL ===\n{scenario['pareto_optimal']['description']}\n"
        + "\n".join(f"  {t}: {a}" for t, a in scenario["pareto_optimal"].items() if t != "description")
        + f"\n\n=== NEGOTIATION TRANSCRIPT ===\n{convo}\n\n"
        f"=== VALDORIA FINAL TREATY ===\n{v_treaty_text}\n\n"
        f"=== KETHARA FINAL TREATY ===\n{k_treaty_text}"
    )

    all_criteria = (
        scenario["judge_criteria"]["negotiation_quality"]
        + scenario["judge_criteria"]["treaty_completeness"]
    )
    n_quality = len(scenario["judge_criteria"]["negotiation_quality"])
    n_complete = len(scenario["judge_criteria"]["treaty_completeness"])

    details = []

    with kbench.chats.new(f"judge_{scenario['name']}"):
        assessment = kbench.assertions.assess_response_with_judge(
            criteria=all_criteria,
            response_text=judge_text,
            judge_llm=judge_llm,
        )

    quality_passed = 0
    complete_passed = 0

    if assessment and assessment.results:
        for i, r in enumerate(assessment.results):
            dim = "negotiation_quality" if i < n_quality else "treaty_completeness"
            details.append({
                "dim": dim,
                "criterion": r.criterion,
                "passed": r.passed,
                "reason": r.reason if hasattr(r, "reason") and r.reason else "",
            })
            if r.passed:
                if dim == "negotiation_quality":
                    quality_passed += 1
                else:
                    complete_passed += 1
    else:
        for c in all_criteria:
            details.append({
                "dim": "unknown",
                "criterion": c,
                "passed": False,
                "reason": "Judge returned no assessment.",
            })

    quality_score = (quality_passed / n_quality * 0.15) if n_quality > 0 else 0.15
    complete_score = (complete_passed / n_complete * 0.15) if n_complete > 0 else 0.15

    return quality_score, complete_score, details


# --------------------------------------------------------------------------------
# 7. CONVERSATION HISTORY MANAGEMENT
# --------------------------------------------------------------------------------
MAX_HISTORY_CHARS = 12000


def truncate_history(history_lines):
    full = "\n".join(history_lines)
    if len(full) <= MAX_HISTORY_CHARS:
        return full
    while len("\n".join(history_lines)) > MAX_HISTORY_CHARS and len(history_lines) > 1:
        history_lines.pop(0)
    return "\n".join(history_lines)


# --------------------------------------------------------------------------------
# 8. THE BENCHMARK TASK
# --------------------------------------------------------------------------------
@kbench.task(name="Peace Treaty Negotiation")
def treaty_negotiation(llm) -> float:
    judge_llm = kbench.llms[JUDGE_MODEL_NAME]
    scenario_scores = []

    for scenario in SCENARIO_SUITE:
        print(f"\n{'='*70}")
        print(f"  {scenario['name']}  [{scenario['tier']}]")
        print(f"  Turn limit: {scenario['turn_limit']}")
        print(f"{'='*70}")

        v_system = build_nation_prompt(
            "Valdoria", scenario["shared_background"], scenario["valdoria_briefing"]
        )
        k_system = build_nation_prompt(
            "Kethara", scenario["shared_background"], scenario["kethara_briefing"]
        )

        v_history = []
        k_history = []
        all_conversation = []
        v_last_msg = ""
        k_last_msg = ""

        # ── NEGOTIATION PHASE ──
        for turn in range(1, scenario["turn_limit"] + 1):
            print(f"\n--- Turn {turn}/{scenario['turn_limit']} ---")

            # ── Valdoria speaks ──
            v_turn_ctx = f"--- Negotiation round {turn} of {scenario['turn_limit']} ---\n"
            if k_last_msg:
                v_turn_ctx += f"Kethara's negotiator says: {k_last_msg}\n"
            else:
                v_turn_ctx += "You are opening the negotiation. Make your initial proposal.\n"

            v_history.append(v_turn_ctx)

            v_prompt = (
                v_system
                + "\n=== NEGOTIATION SO FAR ===\n"
                + truncate_history(list(v_history))
                + "\n\nRespond with your MESSAGE for this round."
            )

            with kbench.chats.new(f"valdoria_{scenario['name']}_{turn}"):
                v_raw = llm.prompt(v_prompt)
            v_last_msg = parse_message(v_raw)

            v_history.append(f"Your message: {v_last_msg}")
            all_conversation.append(f"Valdoria (Round {turn}): {v_last_msg}")

            print(f"  VALDORIA: {v_last_msg}")

            # ── Kethara speaks ──
            k_turn_ctx = f"--- Negotiation round {turn} of {scenario['turn_limit']} ---\n"
            k_turn_ctx += f"Valdoria's negotiator says: {v_last_msg}\n"

            k_history.append(k_turn_ctx)

            k_prompt = (
                k_system
                + "\n=== NEGOTIATION SO FAR ===\n"
                + truncate_history(list(k_history))
                + "\n\nRespond with your MESSAGE for this round."
            )

            with kbench.chats.new(f"kethara_{scenario['name']}_{turn}"):
                k_raw = llm.prompt(k_prompt)
            k_last_msg = parse_message(k_raw)

            k_history.append(f"Your message: {k_last_msg}")
            all_conversation.append(f"Kethara (Round {turn}): {k_last_msg}")

            print(f"  KETHARA:  {k_last_msg}")

        # ── TREATY PROPOSALS ──
        conversation_text = "\n".join(all_conversation)

        v_treaty_prompt = build_treaty_prompt(
            "Valdoria", scenario["valdoria_briefing"], conversation_text
        )
        with kbench.chats.new(f"valdoria_treaty_{scenario['name']}"):
            v_treaty_raw = llm.prompt(v_treaty_prompt)
        valdoria_treaty = parse_treaty(v_treaty_raw)

        k_treaty_prompt = build_treaty_prompt(
            "Kethara", scenario["kethara_briefing"], conversation_text
        )
        with kbench.chats.new(f"kethara_treaty_{scenario['name']}"):
            k_treaty_raw = llm.prompt(k_treaty_prompt)
        kethara_treaty = parse_treaty(k_treaty_raw)

        print(f"\n{'- '*35}")
        print(f"  FINAL TREATY PROPOSALS")
        print(f"{'- '*35}")
        print(f"  VALDORIA's proposal:")
        for k, v in valdoria_treaty.items():
            print(f"    {k}: {v}")
        print(f"  KETHARA's proposal:")
        for k, v in kethara_treaty.items():
            print(f"    {k}: {v}")

        # ── DETERMINISTIC SCORING ──
        deal_score = score_deal_compatibility(valdoria_treaty, kethara_treaty)
        redline_score = score_red_lines(valdoria_treaty, kethara_treaty, scenario)
        priority_score = score_priority_optimization(valdoria_treaty, kethara_treaty, scenario)

        # ── JUDGE SCORING ──
        quality_score, complete_score, judge_details = score_with_judge(
            judge_llm, scenario, all_conversation, valdoria_treaty, kethara_treaty
        )

        total = deal_score + redline_score + priority_score + quality_score + complete_score
        scenario_scores.append(total)

        # ── Print results ──
        print(f"\n{'- '*35}")
        print(f"  RESULTS  |  Score: {total:.3f}/1.00")
        print(f"  (Judge: {JUDGE_MODEL_NAME})")
        print(f"{'- '*35}")

        print(f"\n  Deal Compatibility:     {deal_score:.3f}/0.20")
        # Show per-territory compatibility
        for terr in TERRITORIES:
            v_a = get_territory_assignment(valdoria_treaty, terr)
            k_a = get_territory_assignment(kethara_treaty, terr)
            match = "MATCH" if v_a.lower() == k_a.lower() else "MISMATCH"
            print(f"    {terr:<16s}  V={v_a:<10s}  K={k_a:<10s}  [{match}]")

        print(f"\n  Red Line Compliance:    {redline_score:.3f}/0.20")
        print(f"  Priority Optimization:  {priority_score:.3f}/0.30")

        # Pareto comparison
        pareto = scenario["pareto_optimal"]
        print(f"    Pareto optimal:")
        for terr in TERRITORIES:
            p_assign = pareto.get(terr, "?")
            v_assign = get_territory_assignment(valdoria_treaty, terr)
            match = "=" if v_assign.lower() == p_assign.lower() else "x"
            print(f"      {terr:<16s}  optimal={p_assign:<10s}  actual={v_assign:<10s}  [{match}]")

        print(f"\n  Negotiation Quality:    {quality_score:.3f}/0.15  (judge)")
        print(f"  Treaty Completeness:    {complete_score:.3f}/0.15  (judge)")

        for detail in judge_details:
            icon = "PASS" if detail["passed"] else "FAIL"
            print(f"    [{icon}] {detail['criterion']}")
            if detail["reason"]:
                print(f"            {detail['reason'][:200]}")

    # ── Final Score ──
    final_score = sum(scenario_scores) / len(SCENARIO_SUITE)

    print(f"\n{'='*70}")
    print(f"  PEACE TREATY NEGOTIATION")
    print(f"  FINAL SCORE: {final_score:.4f}")
    print(f"  Per-scenario: [{', '.join(f'{s:.3f}' for s in scenario_scores)}]")
    print(f"  Judge: {JUDGE_MODEL_NAME}")
    print(f"{'='*70}")

    return final_score


# --------------------------------------------------------------------------------
# 9. EXECUTION
# --------------------------------------------------------------------------------
treaty_negotiation.run(kbench.llm)
```
