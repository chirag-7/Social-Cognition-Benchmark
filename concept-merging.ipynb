```python
import kaggle_benchmarks as kbench
from IPython.display import HTML, display

# ================================================================================
# CONCEPT MERGING / COLLABORATIVE DEFINITION BENCHMARK
# Two instances of the same LLM cooperate under asymmetric information.
# Player 1 has half the attributes of a hidden concept.
# Player 2 has the other half.
# They must communicate to synthesize the concept without premature guessing.
# ================================================================================

# --------------------------------------------------------------------------------
# 1. THE DATASET (CONCEPT SUITE) - 25 Levels of Escalating Difficulty
# --------------------------------------------------------------------------------
CONCEPT_SUITE = [
    # --- TIER 1: EASY (Standard Definitions) ---
    {
        "name": "Concept 1",
        "target": "Photosynthesis",
        "p1_attributes": "A biological process that occurs in specialized organelles containing a green pigment. It consumes carbon dioxide.",
        "p2_attributes": "The primary natural mechanism for converting electromagnetic radiation (light) into chemical energy. It produces oxygen as a byproduct.",
        "limit": 6
    },
    {
        "name": "Concept 2",
        "target": "Black Hole",
        "p1_attributes": "An astronomical region of spacetime where the gravitational pull is extremely intense.",
        "p2_attributes": "A physical phenomenon bounded by an 'event horizon' from which not even light can escape.",
        "limit": 6
    },
    {
        "name": "Concept 3",
        "target": "Democracy",
        "p1_attributes": "A system of government where power is vested in the people, who rule either directly or through freely elected representatives.",
        "p2_attributes": "A political structure characterized by majority rule, protection of minority rights, and the guarantee of fundamental human rights.",
        "limit": 6
    },
    {
        "name": "Concept 4",
        "target": "Natural Selection",
        "p1_attributes": "A mechanism of evolution formulated by Charles Darwin, often summarized as 'survival of the fittest'.",
        "p2_attributes": "The differential survival and reproduction of individuals due to differences in phenotype, acting over generations.",
        "limit": 6
    },
    {
        "name": "Concept 5",
        "target": "Inflation",
        "p1_attributes": "An economic concept characterized by a general increase in prices and fall in the purchasing value of money.",
        "p2_attributes": "A metric often measured by the Consumer Price Index (CPI), reflecting a decrease in the currency's buying power over time.",
        "limit": 6
    },

    # --- TIER 2: MEDIUM (Abstract Synthesis) ---
    {
        "name": "Concept 6",
        "target": "Blockchain",
        "p1_attributes": "A growing, sequential list of digital records, securely linked together using cryptographic hashes.",
        "p2_attributes": "A decentralized, distributed, and public digital ledger used to record transactions across many computers immutably.",
        "limit": 8
    },
    {
        "name": "Concept 7",
        "target": "Placebo Effect",
        "p1_attributes": "A phenomenon where a person experiences a perceived improvement in their condition despite receiving a biologically inert treatment.",
        "p2_attributes": "A beneficial outcome resulting purely from a person's belief in or expectation of a treatment's efficacy.",
        "limit": 8
    },
    {
        "name": "Concept 8",
        "target": "Quantum Entanglement",
        "p1_attributes": "A physical phenomenon where groups of particles are generated or interact in ways such that the quantum state of each particle cannot be described independently.",
        "p2_attributes": "A state where the measurement of a particle's physical properties instantaneously dictates the properties of its partner, regardless of the distance separating them.",
        "limit": 8
    },
    {
        "name": "Concept 9",
        "target": "Cognitive Dissonance",
        "p1_attributes": "A psychological term referring to the mental discomfort experienced by someone holding two or more contradictory beliefs, ideas, or values.",
        "p2_attributes": "A state of psychological stress that occurs when a person's behavior is inconsistent with their thoughts and beliefs, often leading to rationalization.",
        "limit": 8
    },
    {
        "name": "Concept 10",
        "target": "Moore's Law",
        "p1_attributes": "An observation and projection in the history of computing hardware, originally made by a co-founder of Intel in 1965.",
        "p2_attributes": "The empirical rule that the number of transistors in a dense integrated circuit doubles about every two years.",
        "limit": 8
    },

    # --- TIER 3: HARD (Specific Theories & Overlapping Attributes) ---
    {
        "name": "Concept 11",
        "target": "CRISPR-Cas9",
        "p1_attributes": "A family of DNA sequences found in the genomes of prokaryotic organisms such as bacteria and archaea, used to detect and destroy DNA from similar bacteriophages.",
        "p2_attributes": "A technology adapted into a powerful tool for genome editing, utilizing a specific enzyme to act as molecular scissors.",
        "limit": 10
    },
    {
        "name": "Concept 12",
        "target": "Tragedy of the Commons",
        "p1_attributes": "A situation where individuals with access to a shared resource act in their own short-term self-interest.",
        "p2_attributes": "An economic and sociological problem where uncoordinated individual action ultimately depletes a shared resource, contrary to the common good of all users.",
        "limit": 10
    },
    {
        "name": "Concept 13",
        "target": "Fractional Reserve Banking",
        "p1_attributes": "A system where financial institutions are required to hold only a portion of the funds deposited by their customers as liquid assets.",
        "p2_attributes": "The mechanism by which most modern fiat money is created, allowing banks to lend out the majority of deposited capital to stimulate economic growth.",
        "limit": 10
    },
    {
        "name": "Concept 14",
        "target": "Dark Matter",
        "p1_attributes": "A hypothetical form of mass thought to account for approximately 85% of the matter in the universe, which does not interact with the electromagnetic field.",
        "p2_attributes": "An unseen substance whose existence is implied by gravitational effects on visible matter, such as anomalies in the rotational speeds of galaxies.",
        "limit": 10
    },
    {
        "name": "Concept 15",
        "target": "Confirmation Bias",
        "p1_attributes": "The tendency to search for, interpret, favor, and recall information in a way that supports one's prior beliefs or values.",
        "p2_attributes": "A systematic error of inductive reasoning where people selectively gather empirical evidence while ignoring or severely undervaluing contradictory data.",
        "limit": 10
    },

    # --- TIER 4: VERY HARD (Complex Interactions & Nuance) ---
    {
        "name": "Concept 16",
        "target": "Ship of Theseus",
        "p1_attributes": "A thought experiment concerning whether an object that has had all of its components replaced remains fundamentally the same object.",
        "p2_attributes": "A philosophical paradox popularized by Plutarch, raising questions about identity, material constitution, and temporal continuity.",
        "limit": 12
    },
    {
        "name": "Concept 17",
        "target": "Bose-Einstein Condensate",
        "p1_attributes": "A state of matter consisting of a dilute gas of bosons cooled to temperatures very close to absolute zero (-273.15 °C).",
        "p2_attributes": "A macroscopic quantum phenomenon where a large fraction of particles occupy the lowest quantum state, making quantum effects apparent on a large scale.",
        "limit": 12
    },
    {
        "name": "Concept 18",
        "target": "Nash Equilibrium",
        "p1_attributes": "A solution concept in non-cooperative games involving two or more players where no player has anything to gain by changing only their own strategy.",
        "p2_attributes": "A state in game theory where each participant is assumed to know the strategies of all other participants, resulting in a mutually locked, optimal holding pattern.",
        "limit": 12
    },
    {
        "name": "Concept 19",
        "target": "Epigenetics",
        "p1_attributes": "The study of heritable changes in gene expression or cellular phenotype that do not involve alterations to the underlying DNA sequence.",
        "p2_attributes": "Biological mechanisms like DNA methylation and histone modification that turn genes 'on' or 'off' in response to environmental factors or age.",
        "limit": 12
    },
    {
        "name": "Concept 20",
        "target": "Keynesian Multiplier",
        "p1_attributes": "An economic theory predicting that an initial injection of autonomous government spending will result in a disproportionately larger increase in national income.",
        "p2_attributes": "A ratio driven by the marginal propensity to consume (MPC), demonstrating how one person's spending becomes another person's income in a continuous cycle.",
        "limit": 12
    },

    # --- TIER 5: INSANELY HARD / ADVERSARIAL (Trap-Door Concepts) ---
    # These are designed to make the LLMs jump to the wrong conclusion if they don't explicitly merge the clues.
    {
        "name": "Concept 21: The Economic Trap",
        "target": "Giffen Good",
        # P1 will think it's a "Veblen Good" (luxury status symbol).
        "p1_attributes": "A highly unusual product characterized by an upward-sloping demand curve, where an increase in price paradoxically leads to an increase in demand.",
        # P2 will think it's a generic "Inferior Good" (like cheap ramen).
        "p2_attributes": "A non-luxury, deeply inferior product with no close substitutes. For this good, the 'income effect' completely overwhelms the 'substitution effect'.",
        "limit": 15
    },
    {
        "name": "Concept 22: The Physics Trap",
        "target": "Casimir Effect",
        # P1 will think it's "Gravity" or "Magnetism".
        "p1_attributes": "A physical force arising from a quantized field, manifesting as a measurable macroscopic attraction between objects that have no distinct electrical charge or mass-based pull.",
        # P2 will think it's "Quantum Tunneling" or "Hawking Radiation".
        "p2_attributes": "A phenomenon occurring in a vacuum, where the presence of boundaries restricts the allowed wavelengths of virtual particles, causing a difference in vacuum energy density.",
        "limit": 15
    },
    {
        "name": "Concept 23: The Mathematical Logic Trap",
        "target": "Gödel's First Incompleteness Theorem",
        # P1 will think it's "Heisenberg's Uncertainty Principle" or "Chaos Theory".
        "p1_attributes": "A foundational principle stating that within a sufficiently complex, effectively generated system, there are fundamental limits to what can be both consistent and absolute.",
        # P2 will think it's a "Paradox" or "Axiom of Choice".
        "p2_attributes": "The mathematical proof demonstrating that in any system capable of basic arithmetic, there will always exist statements that are definitively true, but impossible to prove from within that system.",
        "limit": 15
    },
    {
        "name": "Concept 24: The Psychological Trap",
        "target": "Gell-Mann Amnesia Effect",
        # P1 will think it's the "Halo Effect" or "Authority Bias".
        "p1_attributes": "A cognitive phenomenon where an individual reads a source of information and inherently trusts its credibility, assuming it to be highly accurate on a subject they know nothing about.",
        # P2 will think it's the "Dunning-Kruger Effect" or "Imposter Syndrome".
        "p2_attributes": "The experience of being an absolute expert in a specific domain, reading a media article about your domain, and realizing the article is filled with glaring, fundamental errors and misunderstandings.",
        "limit": 15
    },
    {
        "name": "Concept 25: The Information Theory Trap",
        "target": "Collingridge Dilemma",
        # P1 will think it's "The Butterfly Effect" or "The Law of Unintended Consequences".
        "p1_attributes": "A methodological paradox related to innovation: by the time the negative consequences of a new development become apparent and measurable, it is deeply entrenched in society.",
        # P2 will think it's "Analysis Paralysis" or the "Precautionary Principle".
        "p2_attributes": "An issue of control versus information: when a system is in its early stages, it can be easily changed, but we lack the data to know what to change. Once we have the data, the system is too expensive to change.",
        "limit": 15
    }
]

# --------------------------------------------------------------------------------
# 2. SYSTEM PROMPTS
# --------------------------------------------------------------------------------
SYSTEM_PROMPT_TEMPLATE = """You are Player {player_id} in a cooperative Concept Merging game.
You and your partner are trying to identify a hidden target concept.
Neither of you has the full picture. You ONLY know half of the defining attributes.

YOUR SECRET ATTRIBUTES:
"{attributes}"

YOUR GOAL: 
Communicate with your partner, share your attributes, synthesize your combined clues, and guess the exact concept in as few turns as possible.

RULES:
1. Do NOT guess the final concept immediately. Your clues are insufficient on their own, and a wrong guess ends the game in failure.
2. Share your clues clearly and ask your partner for theirs.
3. Once you both have shared and discussed, if you are highly confident, make a guess.
4. If your partner makes a guess, evaluate it in your THOUGHT section.

You MUST respond in EXACTLY this format (three distinct sections):
THOUGHT: <your internal reasoning, analysis of the partner's message, and confidence level>
MESSAGE: <what you say to your partner>
GUESS: <Leave blank if unsure. If highly confident, output the single concept name here>
"""

# --------------------------------------------------------------------------------
# 3. HELPER FUNCTIONS
# --------------------------------------------------------------------------------
def parse_response(raw_text):
    """Extract THOUGHT, MESSAGE, and GUESS from the model's raw output."""
    thought = ""
    message = raw_text
    guess = ""

    try:
        if "THOUGHT:" in raw_text:
            parts = raw_text.split("THOUGHT:")[1]
            if "MESSAGE:" in parts:
                thought = parts.split("MESSAGE:")[0].strip()
                message_part = parts.split("MESSAGE:")[1]
                
                if "GUESS:" in message_part:
                    message = message_part.split("GUESS:")[0].strip()
                    guess_raw = message_part.split("GUESS:")[1].strip()
                    if guess_raw.lower() not in ["", "none", "n/a", "null"]:
                        guess = guess_raw
                else:
                    message = message_part.strip()
    except Exception:
        pass # Fallback to raw text as message if parsing fails

    return thought, message, guess

def evaluate_guess_with_judge(guess, target_concept, judge_llm):
    """Use the Judge LLM to determine if the guess conceptually matches the target."""
    if not guess:
        return False
        
    prompt = (
        f"You are an objective judge.\n"
        f"Target Concept: '{target_concept}'\n"
        f"Player Guess: '{guess}'\n\n"
        f"Does the player's guess accurately and conceptually match the target concept? "
        f"It does not need to be an exact string match, but it must be the exact same underlying concept. "
        f"Respond with ONLY the word 'YES' or 'NO'."
    )
    
    # We use kbench.chats.new to isolate the judge
    with kbench.chats.new("judge_eval"):
        eval_result = judge_llm.prompt(prompt).strip().upper()
        
    return "YES" in eval_result

MAX_HISTORY_CHARS = 10000

def truncate_history(history_lines):
    """Keep the most recent turns if history exceeds the character budget."""
    full = "\n".join(history_lines)
    if len(full) <= MAX_HISTORY_CHARS:
        return full
    while len("\n".join(history_lines)) > MAX_HISTORY_CHARS and len(history_lines) > 1:
        history_lines.pop(0)
    return "\n".join(history_lines)

# --------------------------------------------------------------------------------
# 4. UI RENDERER
# --------------------------------------------------------------------------------
def render_ui(concept_data, turn, p1_thought, p1_msg, p2_thought, p2_msg, made_by=None, guess_text="", status="playing"):
    # Colors
    bg_color = "#0d1117"
    p1_color = "#58a6ff"
    p2_color = "#d29922"
    
    if status == "solved":
        bg_color = "#0a1f0a"
    elif status == "failed":
        bg_color = "#1f0a0a"

    def format_bubble(player, color, thought, msg):
        if not msg:
            return ""
        return f"""
        <div style="margin-bottom: 12px; border-left: 3px solid {color}; padding-left: 10px;">
            <div style="font-weight: bold; color: {color}; font-size: 12px;">{player}</div>
            <details style="font-size: 11px; color: #8b949e; margin-bottom: 4px; cursor: pointer;">
                <summary>View Internal Thought</summary>
                <div style="padding: 4px; background: #161b22; border-radius: 4px; margin-top: 4px;">{thought}</div>
            </details>
            <div style="color: #c9d1d9; font-size: 13px; line-height: 1.4;">{msg}</div>
        </div>
        """

    chat_html = ""
    if p1_msg: chat_html += format_bubble("PLAYER 1", p1_color, p1_thought, p1_msg)
    if p2_msg: chat_html += format_bubble("PLAYER 2", p2_color, p2_thought, p2_msg)

    # Resolution Banner
    resolution_html = ""
    if guess_text:
        res_color = "#3fb950" if status == "solved" else "#f85149"
        res_icon = "✅" if status == "solved" else "❌"
        resolution_html = f"""
        <div style="background: #21262d; border: 1px solid {res_color}; padding: 10px; border-radius: 6px; margin-top: 12px; text-align: center;">
            <div style="color: {res_color}; font-weight: bold;">{res_icon} {made_by} Guessed: "{guess_text}"</div>
            <div style="color: #8b949e; font-size: 12px; margin-top: 4px;">Target was: {concept_data['target']}</div>
        </div>
        """

    html = f"""
    <div style="background:{bg_color}; padding:16px; border-radius:10px; border:1px solid #30363d; font-family:'Segoe UI',system-ui,sans-serif; max-width:800px; margin-bottom: 16px;">
        <div style="display:flex; justify-content:space-between; align-items:center; border-bottom: 1px solid #30363d; padding-bottom: 8px; margin-bottom: 12px;">
            <span style="font-weight:600; font-size:16px; color:#f0f6fc;">{concept_data['name']}</span>
            <span style="font-size:13px; color:#8b949e;">Turn {turn} / {concept_data['limit']}</span>
        </div>
        
        <div style="display:flex; gap:12px; margin-bottom: 16px;">
            <div style="flex:1; background:#161b22; border: 1px solid #21262d; padding: 10px; border-radius: 6px;">
                <div style="color:{p1_color}; font-size: 11px; font-weight: bold; text-transform: uppercase;">P1 Secret Attributes</div>
                <div style="color:#c9d1d9; font-size: 12px; margin-top: 4px;">{concept_data['p1_attributes']}</div>
            </div>
            <div style="flex:1; background:#161b22; border: 1px solid #21262d; padding: 10px; border-radius: 6px;">
                <div style="color:{p2_color}; font-size: 11px; font-weight: bold; text-transform: uppercase;">P2 Secret Attributes</div>
                <div style="color:#c9d1d9; font-size: 12px; margin-top: 4px;">{concept_data['p2_attributes']}</div>
            </div>
        </div>
        
        <div style="background: #0d1117; border: 1px solid #30363d; padding: 12px; border-radius: 6px;">
            <div style="color: #8b949e; font-size: 11px; margin-bottom: 8px; text-transform: uppercase;">Conversation Timeline</div>
            {chat_html}
        </div>
        
        {resolution_html}
    </div>
    """
    return html

# --------------------------------------------------------------------------------
# 5. THE BENCHMARK TASK
# --------------------------------------------------------------------------------
@kbench.task(name="concept_merging_coop", description="Tests collaborative reasoning and uncertainty signaling between two LLM instances.")
def concept_merging_benchmark(llm) -> float:
    game_scores = []

    for concept in CONCEPT_SUITE:
        p1_system = SYSTEM_PROMPT_TEMPLATE.format(player_id="1", attributes=concept["p1_attributes"])
        p2_system = SYSTEM_PROMPT_TEMPLATE.format(player_id="2", attributes=concept["p2_attributes"])
        
        shared_history = []
        solved = False
        final_status = "playing"
        
        p1_thought, p1_msg = "", ""
        p2_thought, p2_msg = "", ""
        guesser = None
        guess_text = ""

        for turn in range(1, concept["limit"] + 1):
            
            # --- PLAYER 1 TURN ---
            p1_prompt = p1_system + "\n\n=== CONVERSATION HISTORY ===\n" + (truncate_history(shared_history) if shared_history else "No messages yet. You start.") + "\n\nYour turn. Output THOUGHT, MESSAGE, and GUESS."
            
            with kbench.chats.new(f"p1_{concept['name']}_{turn}"):
                p1_raw = llm.prompt(p1_prompt)
            
            p1_thought, p1_msg, p1_guess = parse_response(p1_raw)
            shared_history.append(f"Player 1: {p1_msg}")

            if p1_guess:
                guesser, guess_text = "Player 1", p1_guess
                solved = evaluate_guess_with_judge(p1_guess, concept["target"], kbench.judge_llm)
                final_status = "solved" if solved else "failed"
                display(HTML(render_ui(concept, turn, p1_thought, p1_msg, "", "", guesser, guess_text, final_status)))
                break

            # --- PLAYER 2 TURN ---
            p2_prompt = p2_system + "\n\n=== CONVERSATION HISTORY ===\n" + truncate_history(shared_history) + "\n\nYour turn. Output THOUGHT, MESSAGE, and GUESS."
            
            with kbench.chats.new(f"p2_{concept['name']}_{turn}"):
                p2_raw = llm.prompt(p2_prompt)
                
            p2_thought, p2_msg, p2_guess = parse_response(p2_raw)
            shared_history.append(f"Player 2: {p2_msg}")

            if p2_guess:
                guesser, guess_text = "Player 2", p2_guess
                solved = evaluate_guess_with_judge(p2_guess, concept["target"], kbench.judge_llm)
                final_status = "solved" if solved else "failed"
                display(HTML(render_ui(concept, turn, p1_thought, p1_msg, p2_thought, p2_msg, guesser, guess_text, final_status)))
                break
                
            # Render intermediate state if no guess made
            display(HTML(render_ui(concept, turn, p1_thought, p1_msg, p2_thought, p2_msg)))

        # Loop ended. Calculate Score
        if solved:
            # Score scales linearly. E.g., limit 8. Turn 1/2 = 1.0, Turn 8 = 0.5. 
            # Efficiency matters, but solving it at all guarantees at least 50%.
            efficiency_penalty = (turn - 1) / max(1, (concept["limit"] - 1))
            score = 1.0 - (efficiency_penalty * 0.5)
            game_scores.append(score)
        else:
            if final_status == "playing":
                # Hit turn limit without guessing
                display(HTML(render_ui(concept, turn, p1_thought, p1_msg, p2_thought, p2_msg, "System", "TIMED OUT", "failed")))
            game_scores.append(0.0)

    # --- FINAL SCORE REPORT ---
    final_score = sum(game_scores) / len(CONCEPT_SUITE)

    display(HTML(
        f'<div style="background:#0d1117; padding:20px; border-radius:10px; border:1px solid #a371f7;'
        f'font-family:monospace; color:#f0f6fc; text-align:center; margin:16px 0; max-width:800px;">'
        f'<div style="font-size:13px; color:#8b949e; margin-bottom:4px;">CONCEPT MERGING BENCHMARK</div>'
        f'<div style="font-size:36px; font-weight:700; color:#a371f7;">{final_score:.4f}</div>'
        f'<div style="font-size:13px; color:#8b949e; margin-top:4px;">Final Score (0.0 – 1.0)</div>'
        f'<div style="font-size:12px; color:#484f58; margin-top:12px; padding-top:12px; border-top:1px solid #21262d;">'
        f'{sum(1 for s in game_scores if s > 0)}/{len(CONCEPT_SUITE)} concepts solved · '
        f'Individual Scores: [{", ".join(f"{s:.2f}" for s in game_scores)}]</div>'
        f'</div>'
    ))

    return final_score

# --------------------------------------------------------------------------------
# 6. EXECUTION
# --------------------------------------------------------------------------------
concept_merging_benchmark.run(kbench.llm)
```
